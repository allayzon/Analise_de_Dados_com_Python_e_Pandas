# **Dicionários**

```python
dc = {"Maçã":20, "Banana":10, "Laranja":15, "Uva":5}  # Dicts trabalham com chave-valor
dc
```

```python
dc['Maçã']  # Chave: Maçã, Valor: 20
```

```python
dc.keys()  # Retorna todas as chaves
```

```python
dc.values()  # Retorna todos os valores
```

```python
dc['Maçã'] = 25  # Dicionários são mutáveis, essa linha vai rodar normalmente
dc['Maçã']
```

```python
dc.setdefault("Limão", 22)  # Verifica se existe Limão e retorna o mesmo, se não existir, será inserido a chave limão com o valor 22
dc
```
